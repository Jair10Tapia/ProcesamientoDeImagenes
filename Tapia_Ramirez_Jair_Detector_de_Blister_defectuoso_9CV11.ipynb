```python
import cv2
import imutils
from google.colab.patches import cv2_imshow

img = cv2.imread("/content/Blister-lleno.jpeg") #Lectura de imagen
img = imutils.resize(img, width=400) 
grises = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #Cambio de modelo de color. RGB a Escala de grises
grises = cv2.GaussianBlur(grises, (23, 23), 0) #Aplicación de filtro Gaussiano
umbral = cv2.threshold(grises, 160, 255, cv2.THRESH_BINARY)[1] #Umbralización de la imagen
contador = cv2.findContours(umbral, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)  #Función para encontrar contornos dentro de la imagen. 
contador = imutils.grab_contours(contador)
salidablister = img.copy()

print("Imagen en Escala de grises") #Impresión de imagen en escala de grises con filtro Gaussiano
cv2_imshow(grises)
print("\n")
print("Imagen con Umbralización") #Impresion de Imagen umbralizada
cv2_imshow(umbral)

#for de contador para los contornos de las imagenes encontradas.
for c in contador:
    if cv2.contourArea(c) < 900:
        continue
    (x, y, w, h) = cv2.boundingRect(c)
    cv2.rectangle(salidablister, (x, y), (x + w, y + h), (0, 0, 255), 3)
    faltan = str(10 - len(contador))


print("\n")
print ("Detección del contenido del blister")
#Condicionales para los elementos detectados previamente
if len(contador) < 10:
    text = "Solo hay {} pastillas".format(len(contador)) + " y faltan " + faltan + " pastillas"
    print (text)
    cv2.putText(salidablister, text, (10,25), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0),2)
if len(contador) == 10:
    text = "Blister completo 'aprobado' "
    print(text)
    cv2.putText(salidablister, text, (10,25), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (43, 255, 0),2)

cv2_imshow(salidablister) #impresión de la salida con los rectangulos en cada elemento encontrado. 
```
