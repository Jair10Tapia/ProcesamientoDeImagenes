Identidad

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
imagen = cv2.imread("img1.jpg")
height, width = imagen.shape[0:2]

Identidad=[[0 for i in range(width)]for j in range(height)]

for i in range(height):
    for j in range(width):
        Identidad[i][j]=imagen[i][j]

Identidad=np.uint8(Identidad)
cv2_imshow(Identidad)
```

Inverso Negativo

```python
imagen = cv2.imread("img1.jpg")
height, width = imagen.shape[0:2]

Negativo=[[0 for i in range(width)]for j in range(height)]

for i in range(height):
    for j in range(width):
        inverso=np.array(imagen[i][j])
        Negativo[i][j]=255-inverso

Negativo=np.uint8(Negativo)
cv2_imshow(Negativo)
```

Umbral

```python
imagen = cv2.imread("img1.jpg")
height, width = imagen.shape[0:2]

Umbral=[[0 for i in range(width)]for j in range(height)]

p1=40    
p2=110

#Dependiendo los valores, es la visualizacion de la imagen.

for i in range(height):
    for j in range(width):
        if (imagen[i][j][1]<=p1 or imagen[i][j][1] >=p2):
            Umbral[i][j]=255
        else:
            Umbral[i][j]=0

Umbral=np.uint8(Umbral)
cv2_imshow(Umbral)
```

```python
imagen = cv2.imread("img1.jpg")

height, width = imagen.shape[0:2]

Binario=[[0 for i in range(width)]for j in range(height)]

p1=40       
p2=110

#Dependiendo los valores, se visualiza mejor la imagen.

for i in range(height):
    for j in range(width):
        if (imagen[i][j][1] < p1 or imagen[i][j][1] >=p2):
            Binario[i][j]=255
        else:
            Binario[i][j]=0

Binario=np.uint8(Binario)
cv2_imshow(Binario)
```

```python
imagen = cv2.imread("img1.jpg")
height, width = imagen.shape[0:2]
binarioinvertido=[[0 for i in range(width)]for j in range(height)]

p1=40       #Entre mas altos estos valores, mejor visualizacion de la imagen 
p2=110

for i in range(height):
    for j in range(width):
        if (imagen[i][j][1] <= p1 or imagen[i][j][1]>=p2):
            binarioinvertido[i][j]=0
        else:
            binarioinvertido[i][j]=255

binarioinvertido=np.uint8(binarioinvertido)
cv2_imshow(binarioinvertido)
```

Umbral de escala de grises

```python
imagen = cv2.imread("img1.jpg")
height, width = imagen.shape[0:2]
UmbralGrises=[[0 for i in range(width)]for j in range(height)]
P1=40
P2=110
for i in range(height):
    for j in range(width):
        if (imagen[i][j][1] <= P1 or imagen[i][j][1] >=P2):
            UmbralGrises[i][j]=255
        else:
            UmbralGrises[i][j]=imagen[i][j][1]

UmbralGrises=np.uint8(UmbralGrises)
cv2_imshow(UmbralGrises)
```

```python
import matplotlib.pyplot as plt

imagen = cv2.imread("img1.jpg")

height, width = imagen.shape[0:2]
listaDeValores=[]

for i in range(height):
    for j in range(width):
        variable=np.array(imagen[i][j][0])
        listaDeValores.append(variable)


diccionarioContador = {}

for contador in listaDeValores:
    contador=int(contador)
    if(contador in diccionarioContador):
        diccionarioContador[contador]+=1
    else:
        diccionarioContador[contador]=1

print(diccionarioContador)
valorX=[]

for valor in diccionarioContador:
    agregar=diccionarioContador[valor]
    print(agregar)
    valorX.append(agregar)

print(valorX)

plt.stem(valorX)
plt.show() 
    
```
