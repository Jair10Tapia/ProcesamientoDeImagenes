Logar칤tmica

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)
B, G, R = cv2.split(imagen)
BP=(255/np.log(1.2*255+1))*np.log(1.2*B+1)
GP=(255/np.log(1.2*255+1))*np.log(1.2*G+1)
RP=(255/np.log(1.2*255+1))*np.log(1.2*R+1)
RES = cv2.merge([BP,GP,RP])
RES=np.uint8(RES)
cv2_imshow(RES)
cv2.waitKey(0)
```

Funci칩n Senoidal

```python
import math
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)
B, G, R = cv2.split(imagen) #Dividir en RGB
BP=255*np.sin(((math.pi)*B)/(2*255))
GP=255*np.sin(((math.pi)*G)/(2*255))
RP=255*np.sin(((math.pi)*R)/(2*255))
RES = cv2.merge([BP,GP,RP])
RES=np.uint8(RES)
cv2_imshow(RES)
cv2.waitKey(0)
```

Funci칩n Exponencial

```python
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)
B, G, R = cv2.split(imagen)
BP=(255/(1-(np.exp(-1.3))))*(1-np.exp((-1.3*B)/255))
GP=(255/(1-(np.exp(-1.3))))*(1-np.exp((-1.3*G)/255))
RP=(255/(1-(np.exp(-1.3))))*(1-np.exp((-1.3*R)/255))
RES = cv2.merge([BP,GP,RP])
RES=np.uint8(RES)
cv2_imshow(RES)
```

Funci칩n Cosenoidal

```python
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)

height, width = imagen.shape[0:2]

coseinoidal=[[0 for i in range(width)]for j in range(height)]

for i in range(height):
    for j in range(width):
        coseinoidal[i][j]=255*(1-(np.cos((np.pi*(imagen[i,j]))/(2*255))))

coseinoidal=np.uint8(coseinoidal)

cv2_imshow(coseinoidal)
```

Exponencial Creciente

```python
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)

height, width = imagen.shape[0:2]

creciente=[[0 for i in range(width)]for j in range(height)]

alpha=3
constante=(255/((np.e**alpha)-1))

for i in range(height):
 for j in range(width):
   creciente[i][j]=constante*(np.e**((alpha*(imagen[i,j]))/255)-1)

creciente=np.uint8(creciente)

cv2_imshow(creciente)
```

Gamma

```python
imagen= cv2.imread('img1.jpg')
imagen=np.float32(imagen)
B, G, R = cv2.split(imagen)
BP=pow((B/255),0.2)*255
GP=pow((G/255),0.2)*255
RP=pow((R/255),0.2)*255
RES = cv2.merge([BP,GP,RP])
RES=np.uint8(RES)
cv2_imshow(RES)
```
