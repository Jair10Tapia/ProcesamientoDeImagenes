```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

imagen = cv2.imread('img1.jpg')
imagenr = imagen

(B, G, R) = imagenr[:, :, 0], imagenr[:, :, 1], imagenr[:, :, 2]
imagenSG = (R * 0.299) + (G * 0.587) + (B * 0.114)

cv2.imwrite('OriginalGS.jpg', imagenSG)

imagenSGR = cv2.imread('OriginalGS.jpg')
derivadax = np.zeros(imagenr.shape[:2], dtype="uint8")
derivaday = np.zeros(imagenr.shape[:2], dtype="uint8")

size = imagenr.shape[:2]

for x in range(size[0]):

    for y in range(size[1]):

        if(x < size[0] - 1):

            if ((imagenSG[x + 1][y] - imagenSG[x][y]) < 0):

                derivadax[x][y] = 0

            else:

                derivadax[x][y] = imagenSG[x + 1][y] - imagenSG[x][y]

        if(y < size[1] - 1):

            if ((imagenSG[x][y + 1] - imagenSG[x][y]) < 0):

                derivaday[x][y] = 0

            else:

                derivaday[x][y] = imagenSG[x][y + 1] - imagenSG[x][y]

derivadax2 = np.zeros(imagenr.shape[:2], dtype="uint8")
derivaday2 = np.zeros(imagenr.shape[:2], dtype="uint8")

for x in range(size[0]):

    for y in range(size[1]):

        if(x < size[0] - 1):

            if ((imagenSG[x - 1][y]  - (2 * imagenSG[x][y]) + imagenSG[x + 1][y]) < 0):

                derivadax2[x][y] = 0

            else:

                derivadax2[x][y] = imagenSG[x - 1][y]  - (2 * imagenSG[x][y]) + imagenSG[x + 1][y]

        if(y < size[1] - 1):

            if ((imagenSG[x][y - 1]  - (2 * imagenSG[x][y]) + imagenSG[x][y + 1]) < 0):

                derivaday2[x][y] = 0

            else:

                derivaday2[x][y] = imagenSG[x][y - 1]  - (2 * imagenSG[x][y]) + imagenSG[x][y + 1]


cv2_imshow(imagen)
cv2_imshow(imagenSGR)
cv2_imshow(derivadax)
cv2_imshow(derivaday)
cv2_imshow(derivadax2)
cv2_imshow(derivaday2)
```
