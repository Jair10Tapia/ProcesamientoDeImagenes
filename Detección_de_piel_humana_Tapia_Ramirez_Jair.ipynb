Detección de piel humana

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

imagen = cv2.imread('piel.jpg')
size = imagen.shape[:2]

(B, G, R) = imagen[:, :, 0], imagen[:, :, 1], imagen[:, :, 2]

maskR = np.zeros(size, dtype="uint8")
maskG = np.zeros(size, dtype="uint8")
maskB = np.zeros(size, dtype="uint8")
skinR = np.zeros(size, dtype="uint8")
skinG = np.zeros(size, dtype="uint8")
skinB = np.zeros(size, dtype="uint8")

for x in range(size[0]):

    for y in range(size[1]):

        if(R[x][y] > 95 and G[x][y] > 40 and B[x][y] > 20 and (int(max(R[x][y], G[x][y], B[x][y])) - int(min(R[x][y], G[x][y], B[x][y]))) > 15 and abs(int(R[x][y]) - int(G[x][y])) > 15 and R[x][y] > G[x][y] and R[x][y] > B[x][y]):

            maskR[x][y] = 255
            maskG[x][y] = 255
            maskB[x][y] = 255

            skinR[x][y] = R[x][y]
            skinG[x][y] = G[x][y]
            skinB[x][y] = B[x][y]


Mask= cv2.merge([maskB, maskG, maskR])
skin = cv2.merge([skinB, skinG, skinR])

Is2 = cv2.imwrite('Imagenes/Prueba 2/Mascara.jpg', Mask)
Is3 = cv2.imwrite('Imagenes/Prueba 2/Piel.jpg', skin)

cv2_imshow(imagen)
cv2_imshow(Mask)
cv2_imshow(skin)
```
