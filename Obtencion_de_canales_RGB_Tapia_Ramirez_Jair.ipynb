```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
img_name = "img1.jpg"
img_array = cv2.imread(img_name)
h,w,c = img_array.shape
R_img_array = np.zeros ((h, w, 3), dtype = np.uint8) #Inicializa la imagen roja, todos los valores son 0
G_img_array = np.zeros ((h, w, 3), dtype = np.uint8) #Inicializa la imagen verde, todos los valores son 0
B_img_array = np.zeros ((h, w, 3), dtype = np.uint8) #Inicializa la imagen azul, todos los valores son 0

R_img_array[:,:,0] = img_array[:,:,0] 
G_img_array[:,:,1] = img_array[:,:,1]
B_img_array[:,:,2] = img_array[:,:,2]
```

```python
cv2_imshow(img_array)
```
